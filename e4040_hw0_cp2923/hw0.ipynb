```python
import numpy as np
import theano
import theano.tensor as T
from theano import shared
from theano.tensor.shared_randomstreams import RandomStreams
from theano import function
```

```python
x = T.fvector()
```

```python
srng = RandomStreams(seed=1)
a = srng.uniform((10,))
b = srng.uniform((10,))
```

```python
A = shared(a.eval())
B = shared(b.eval())
c = T.dot(T.transpose(x+A), B)
inner = theano.function([x],c)
d =inner(range(10))
```

# Problem 3

```python
def fib(n):
    a = shared(0)
    b = shared(1)
    i = 0
    fib1 = theano.function([],[], updates=[(b,a+b),(a,b)])
    while i < n:
        fib1()
        i = i+1
    return(a.get_value())
```

```python
for i in range(10):
    print(fib(i))
```
